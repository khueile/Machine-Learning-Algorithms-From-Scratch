### Classes and other data structures:
leaf, node

```python
def attri(dataset):
    attributes = {}
    for line in dataset:
        attribute = line[-1]
        if attribute not in attributes:
            attributes[attribute] = 1
        else:
            attributes[attribute] += 1
    return attributes
```

```python
class leaf:

    def __init__(self, dataset):
        self.leaf = attri(dataset)
        

    
```

```python
class node:
    
    def __init__(self, splitbyTuple, nodeT, nodeF):
        self.splitby = splitbyTuple 
        self.nodeT = nodeT
        self.nodeF = nodeF
    
```

### Split
Split a dataset into two smaller datasets by whether a certain column's value in each line is equal to a specific value. Return the smaller categorized dataset.

```python
def split(dataset, col, val):
    Tdata, Fdata = [], []
    for line in dataset:
        if line[col] == val:
            Tdata.append(line)
        else:
            Fdata.append(line)
    return Tdata, Fdata
            
```

### Calculate Entropy of a dataset 
- find the attributes included in current dataset and the number of those restaurants
- calculate entropy


```python
def entropy(dataset):
    
    #fill a dictionary with attributes and their frequency in the dataset
    attributes = {}
    for line in dataset:
        attribute = line[-1]
        if attribute not in attributes:
            attributes[attribute] = float(1/len(dataset))
        else:
            attributes[attribute] += float(1/len(dataset))
    
    #Calculate entropy
    entropy = 0
    for atribute in attributes:
        freq = attributes[attribute]
        entropy -= freq * math.log2(freq)
        
    return entropy
```

### Calulate Information gain for a split
- Calculate impurity of original dataset (ogEn)
- Calculate impurity of splitted datasets (tEn, fEn)
- Calculate rate of lines in dataset that satisfy true dataset over all lines in the original dataset(rate)
- return info gained (ogEn - weighed rate for Tdata and Fdata)

```python
def calcGain(dataset, Tdata, Fdata):
    ogEn = entropy(dataset)
    tEn = entropy(Tdata)
    fEn = entropy(Fdata)
    rate = float(len(Tdata)/len(dataset))
    
    return ogEn - (tEn * rate + fEn * (1-rate))
    
```

### Best way to split
Function takes in a dataset to find the best way to split dataset based on gini impurity and coeficient gain. returns the matching column and value that the best split is based on

- for each column in dataset, 
- for each available values of that column,

+ split the dataset into two smaller datasets Tdata and Fdata
+ calculate the info gain
+ compare gain to best current gain
+ Keep the better gain value and its column and value. 

- return column and value



```python
def bestSplit(dataset):
    maxCol = 0 ###Check accuracy in edge cases! 
    maxVal = 0
    maxGain = 0
    #for each column in dataset
    for col in range(len(dataset[0])-1): 
        
        #make a dictionary containing all possible values in column i of dataset
        colVals = {}
        k=0
        for row in range(len(dataset)):
            rowVal = dataset[row][col]
            if rowVal not in colVals:
                colVals[row] = rowVal
                k+=1
                
        #for each available values of that column
        for valname in range(len(colVals)):
            Tdata, Fdata = split(dataset,col,colVals[valname])
            if (len(Tdata) == 0) or (len(Fdata) == 0):
                continue
            
            gain = calcGain(dataset, Tdata, Fdata)
            if gain > maxGain:
                maxGain = gain
                maxCol = col
                maxVal = colVals[valname]
                
            
            
    return maxCol, maxVal
```

### Recursive function
- Make a node. 
- Choose way to split data
- If there is no way, add leaf
- Split data
- Recursive call for each chunk of dataset splited

```python
def tree(dataset):
    col, val = bestSplit(dataset)
    
    
    Tdataset,Fdataset = split(dataset, col, val)
    #print(len(Fdataset),len(Tdataset))
    
    if len(Fdataset)==0:
        return leaf(Tdataset)
    elif len(Tdataset)==0:
        return leaf(Fdataset)

    
    nodeT = tree(Tdataset)
    nodeF = tree(Fdataset)    
    
    return node((col,val), nodeT, nodeF)
    
```

```python
tree_restaurant = tree(train)
```

```python
def classify(row, node):
    """See the 'rules of recursion' above."""

    # Base case: we've reached a leaf
    if isinstance(node, leaf):
        return node.leaf

    # Decide whether to follow the true-branch or the false-branch.
    # Compare the feature / value stored in the node,
    # to the example we're considering.
    if node.splitby[1] == row[node.splitby[0]] :
        return classify(row, node.nodeT)
    else:
        return classify(row, node.nodeF)
```

```python
correct = 0
for row in test:
    if next(iter(classify(row, tree_restaurant))) == row[-1]:
        correct +=1
accuracy = float(correct/len(test))
print(accuracy)
```
